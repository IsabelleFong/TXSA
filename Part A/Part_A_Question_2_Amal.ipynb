## 1.	Explain the importance of stemming in text analytics 

Stemming is an important text normalization technique in text analytics. The word stem refers can be interpreted to the act of "chopping" a plant to its stem. Likewise stemming in text analytics "chops" common suffixes from a word such as "ing", "ed", "s" and etc. By converting different morphological variations of a word into a common "stem", stemming ensures that related words are treated as the same term during analysis.

Another key point is vocabulary reduction. Text data often contains multiple word that derives from the same root word (e.g.  connect, connected, connecting). Stemming reduces the variations into a single representaion. This decreases feaeture dimensionality and improves overall computation efficiency.

Lastly, in text classification and information retrieval, stemming improves the pattern matching and retrieval performance by grouping similar word forms together. Models can caputre broader patterns in the data without being affected by minor grammatical differences, leading to more consistent analytical results.

## 2.	Demonstrate word stemming using Regular Expression, Porter Stemmer and Lancaster Stemmer and report the output. 

```python
# Import dataset

file_path = "Dataset/Data_1.txt"

with open(file_path, "r") as file:
    text = file.read()

print(text)
```

```python
# Tokenize
import nltk
from nltk.tokenize import word_tokenize

nltk.download('punkt_tab')

word = word_tokenize(text)

print(word)
```

# Porter Stemmer

```python
from nltk.stem import PorterStemmer

ps = PorterStemmer()

ps_stems = [ps.stem(w) for w in word]
```

# Lancaster Stemmer

```python
from nltk.stem import LancasterStemmer

ls = LancasterStemmer()

ls_stems = [ls.stem(w) for w in word]
```

# Regular Expression Stemmer

```python
import re

def regex_stems(word):
    return re.sub(r'(able|ible|ness|ment|ful|less|ious|ous|ish|ive|ity|ty|ist|ism|ation|ition|tion|ion|ing|ly|ed|es|er|s|)$','',word)

regex_stems = [regex_stems(w) for w in word]
```

```python
print("Porter Stemmer\n", ps_stems)
print("\nLancaster Stemmer\n", ls_stems)
print("\nRegular Expression Stemmer\n", regex_stems)
```

## 3.	Explain the differences among Regular Expression stemmer, Porter Stemmer and Lancaster Stemmer using the obtained output.

Porter Stemmer provides a balanced stemming result by using a multi-step algorithm that applies condition-based suffix stripping set of rules in a predefined order. The word "classification" is stemmed to classif, and "considered" becomes "consid". This shows consistent normalization across related word forms and has the least over-stemming. That said the resulted stems are not all valid dictionary words.

Compared to porter, Lancaster Stemmer uses a more aggresive set of rules that removes suffixes without extensive checks. This produces the shortest stems. The word "classification" is stemmed to "class", which is a different word already. The word "isolation" becomes "isol" which is not a valid dictionary word. Thus compared to the other 3, lancaster has the most over-stemmed words.

Unlike Stemmer and  Lancaster, Regular Expression does not rely on linguistic rules, rather it performs directly with pattern matching. This makes Regular Expression Stemmer the most flexible by allowing to manually decide what suffix to be removed. With the suffix I've defined to remove, the word "classification" is stemmed to "classific". However to remove plurals I defined the letter 's' to be removed, this results in the word "class" to be reduced to "clas" which becomes an invalid word, this was not an issue with porter and lancaster.
